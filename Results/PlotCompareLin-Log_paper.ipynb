```python
import numpy as np
import pylab as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.legend_handler import HandlerTuple

# bars and stripes
# lin cost
path_file = 'Data_QNN/linLossBS/ACC_test.txt'
ACC0 = np.genfromtxt(path_file,usecols=0,skip_header=1)
ACC1 = np.genfromtxt(path_file,usecols=1,skip_header=1)
ACC0 = np.array([0.5,*ACC0])
ACC1 = np.array([0.5,*ACC1])
ACCaverLin = (ACC0+ACC1)*0.5
gatesLin=np.linspace(0,len(ACCaverLin)-1,len(ACCaverLin))

#log Cost
path_file = 'Data_QNN/logLossBS/ACC_test.txt'
ACC0 = np.genfromtxt(path_file,usecols=0,skip_header=1)
ACC1 = np.genfromtxt(path_file,usecols=1,skip_header=1)
ACC0 = np.array([0.5,*ACC0])
ACC1 = np.array([0.5,*ACC1])
ACCaverLog = (ACC0+ACC1)*0.5
gatesLog=np.linspace(0,len(ACCaverLog)-1,len(ACCaverLog))

# mnist-pca
# lin cost
path_file = 'Data_QNN/linLossMN/ACC_test.txt'
ACC0 = np.genfromtxt(path_file,usecols=0,skip_header=1)
ACC1 = np.genfromtxt(path_file,usecols=1,skip_header=1)
ACC0 = np.array([0.5,*ACC0])
ACC1 = np.array([0.5,*ACC1])
ACCaverLinMnist = (ACC0+ACC1)*0.5
gatesLinMnist=np.linspace(0,len(ACCaverLinMnist)-1,len(ACCaverLinMnist))

#log Cost
path_file = 'Data_QNN/logLossMN/ACC_test.txt'
ACC0 = np.genfromtxt(path_file,usecols=0,skip_header=1)
ACC1 = np.genfromtxt(path_file,usecols=1,skip_header=1)
ACC0 = np.array([0.5,*ACC0])
ACC1 = np.array([0.5,*ACC1])
ACCaverLogMnist = (ACC0+ACC1)*0.5
gatesLogMnist=np.linspace(0,len(ACCaverLogMnist)-1,len(ACCaverLogMnist))



fig=plt.figure(figsize=(11.69*0.4,8.27*0.4)).gca()

ax1 = plt.subplot(1,1,1)

curve1, = ax1.plot(gatesLog,ACCaverLog,'s-',color='tab:red',label='log cost (b&s)',fillstyle='none')
curve2, = ax1.plot(gatesLin,ACCaverLin,'*-',color='tab:red',label='lin cost (b&s)',fillstyle='none')

curve3, = ax1.plot(gatesLogMnist,ACCaverLogMnist,'d-',color='tab:purple',label='log cost (MNIST)',fillstyle='none')
curve4, = ax1.plot(gatesLinMnist,ACCaverLinMnist,'x-',color='tab:purple',label='lin cost (MNIST)',fillstyle='none')

ax1.set_ylim(0.4,1)

ax1.set_xlim(0,max([max(gatesLin),max(gatesLog)])+0.5)


handles = [(curve1,curve3),(curve2,curve4)]
labels = ['log cost', 'lin cost']
ax1.legend(handles,labels,handler_map={tuple: HandlerTuple(ndivide=None)})#,loc=4)
ax1.set_ylabel('accuracy')
ax1.set_xlabel('number of gates')

fig.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.tight_layout()
plt.savefig('Acc_costLin-Log_paper.pdf')
#plt.close()

```
